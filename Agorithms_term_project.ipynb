```python
import json
import os

FILE_NAME = "student_data.json"

# ==========================================
# File I/O (Storage Management)
# ==========================================
def load_data():
    """Loads student data from the JSON file into a list of dictionaries."""
    if not os.path.exists(FILE_NAME):
        return [] # Return an empty list if the file doesn't exist yet

    with open(FILE_NAME, 'r', encoding='utf-8') as file:
        try:
            return json.load(file)
        except json.JSONDecodeError:
            return [] # Return empty if file is corrupted

def save_data(data):
    """Saves the list of dictionaries back to the JSON file."""
    with open(FILE_NAME, 'w', encoding='utf-8') as file:
        json.dump(data, file, indent=4, ensure_ascii=False)

# ==========================================
# Core Operations (CRUD + Show)
# ==========================================
def add_student(data):
    print("\n--- Add New Student ---")
    student = {
        "Name": input("First Name: ").strip(),
        "Lastname": input("Last Name: ").strip(),
        "Place of Birth": input("Place of Birth: ").strip(),
        "Gender": input("Gender (M/F/Other): ").strip(),
        "Telephone": input("Telephone Number: ").strip(),
        "Emergency Contact": input("Emergency Contact Name: ").strip(),
        "Emergency Phone": input("Emergency Contact Phone: ").strip()
    }
    data.append(student)
    save_data(data)
    print(f"\n✅ Student {student['Name']} added successfully!")

def show_all_students(data):
    """Displays all students currently in the database."""
    print("\n--- All Students List ---")
    if not data:
        print("📭 No students found in the system.")
        return

    print(f"Total students: {len(data)}\n")
    # Topic of the table
    print(f"{'Index':<6} | {'Name':<15} | {'Lastname':<15} | {'Gender':<8} | {'Telephone'}")
    print("-" * 65)

    for idx, student in enumerate(data):
        # ใช้ f-string formatting เพื่อจัดคอลัมน์ให้ตรงกัน (< คือชิดซ้าย, ตัวเลขคือความกว้าง)
        print(f"[{idx:<3}] | {student['Name']:<15} | {student['Lastname']:<15} | {student['Gender']:<8} | {student['Telephone']}")
    print("-" * 65)

def search_students(data):
    print("\n--- Search Students ---")
    print("Search by: 1. Name | 2. Lastname | 3. Gender")
    choice = input("Select an option (1-3): ").strip()

    key_map = {"1": "Name", "2": "Lastname", "3": "Gender"}
    if choice not in key_map:
        print("❌ Invalid choice.")
        return []

    search_key = key_map[choice]
    query = input(f"Enter {search_key} to search for: ").strip().lower()

    results = []
    for index, student in enumerate(data):
        # Case-insensitive partial search
        if query in student.get(search_key, "").lower(): # Changed == to in
            results.append((index, student))

    if not results:
        print("No students found matching that criteria.")
    else:
        print(f"\nFound {len(results)} matching student(s):")
        for idx, student in results:
            print(f"[{idx}] {student['Name']} {student['Lastname']} | Gender: {student['Gender']} | Phone: {student['Telephone']}")

    return results

def update_student(data):
    print("\n--- Update Student ---")
    print("First, search for the student you want to update.")
    results = search_students(data)

    if not results:
        return

    try:
        target_idx = int(input("\nEnter the [Index] number of the student to update: "))
        # Verify the selected index is actually in the search results to prevent accidents
        if target_idx not in [idx for idx, _ in results]:
            print("❌ Invalid index selected.")
            return

        print("\nEnter new values (leave blank to keep current value):")
        student = data[target_idx]
        for key in student.keys():
            new_val = input(f"{key} [{student[key]}]: ").strip()
            if new_val: # Only update if the user typed something
                student[key] = new_val

        save_data(data)
        print("✅ Student updated successfully!")

    except ValueError:
        print("❌ Please enter a valid number.")

def delete_student(data):
    print("\n--- Delete Student ---")
    print("First, search for the student you want to delete.")
    results = search_students(data)

    if not results:
        return

    try:
        target_idx = int(input("\nEnter the [Index] number of the student to delete: "))
        if target_idx not in [idx for idx, _ in results]:
            print("❌ Invalid index selected.")
            return

        confirm = input(f"Are you sure you want to delete {data[target_idx]['Name']}? (y/n): ").strip().lower()
        if confirm == 'y':
            deleted_student = data.pop(target_idx)
            save_data(data)
            print(f"✅ {deleted_student['Name']} has been deleted.")
        else:
            print("Deletion cancelled.")

    except ValueError:
        print("❌ Please enter a valid number.")

# ==========================================
# Main Menu Loop
# ==========================================
def main():
    # Load data into memory once when the program starts
    students_data = load_data()

    while True:
        print("\n" + "="*30)
        print(" STUDENT MANAGEMENT SYSTEM")
        print("="*30)
        print("1. Add Student")
        print("2. Search Student")
        print("3. Update Student")
        print("4. Delete Student")
        print("5. Show All Students")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ").strip()

        if choice == '1':
            add_student(students_data)
        elif choice == '2':
            search_students(students_data)
        elif choice == '3':
            update_student(students_data)
        elif choice == '4':
            delete_student(students_data)
        elif choice == '5':
            show_all_students(students_data) # <--- เรียกใช้ฟังก์ชันที่เพิ่งสร้าง
        elif choice == '6':
            print("Exiting program.")
            break
        else:
            print("❌ Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
